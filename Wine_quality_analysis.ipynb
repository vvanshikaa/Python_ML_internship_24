**Wine Quality Analysis**

You have 2 features as citrus acid content amd density of wine. Calculate the quality by using the formula

citrus acid content can be in range(1 to 5)
density of wine can be in range(9930 to 9970)

quality = 15000*(var1/var2).

Plot the KNN graph and predict the quality of wine

quality range wine status :

5 - 6.5 normal

6.5  - 8 good

8 - 10 best

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
np.random.seed(0)  # For reproducibility
var1 = np.random.uniform(1, 5, 100)  # Citrus acid content
var2 = np.random.uniform(9930, 9970, 100) 
```

```python
quality = 15000 * (var1 / var2)
```

```python
wine_status = []
for q in quality:
    if q < 6.5:
        wine_status.append('normal')
    elif q < 8:
        wine_status.append('good')
    else:
        wine_status.append('best')
```

```python
le = LabelEncoder()
wine_status_encoded = le.fit_transform(wine_status)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(np.column_stack((var1, var2)), wine_status_encoded)
```

```python
plt.scatter(var1, var2, c=wine_status_encoded)
plt.xlabel('Citrus Acid Content')
plt.ylabel('Density of Wine')
plt.title('KNN Graph')
plt.show()
```

```python
new_var1 = 3.5  # Citrus acid content
new_var2 = 9950  # Density of wine
```

```python
new_quality = knn.predict([[new_var1, new_var2]])
new_wine_status = le.inverse_transform(new_quality)[0]
```

```python
print(f'Predicted wine status: {new_wine_status}')
```
